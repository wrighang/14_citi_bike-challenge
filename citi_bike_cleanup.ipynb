#####  This data has been processed to remove trips that are taken by staff as they service and inspect the system, trips that are taken to/from any of our “test” stations (which we were using more in June and July 2013), and any trips that were below 60 seconds in length (potentially false starts or users trying to re-dock a bike to ensure it's secure). 

#### For months with more than 1 million trips, trips data is split into multiple CSVs within the same compressed file. To determine a month’s total rides, sum the records across CSVs.

ride_id: A unique identifier assigned to each ride.

rideable_type: Specifies the type of bike used for the ride.

started_at: The date and time when the ride began.

ended_at: The date and time when the ride concluded.

start_station_name: The name of the station where the ride started.

start_station_id: A unique identifier for the start station.

end_station_name: The name of the station where the ride ended.

end_station_id: A unique identifier for the end station.

start_lat: The latitude coordinate of the start station.

start_lng: The longitude coordinate of the start station.

end_lat: The latitude coordinate of the end station.

end_lng: The longitude coordinate of the end station.

member_casual: Indicates the rider's membership status, distinguishing between annual members and casual riders.

```python
'''validated online before replcing: a "customer" refers to someone who uses the bike sharing service by purchasing a short-term pass like a day pass, while a "subscriber" is someone who has signed up for an annual membership, giving them recurring access to the bikes with typically more benefits like longer ride times and lower per-use costs'''
#replace 'Customer' with 'casual' and 'Subscriber' with 'member' in "Member Casual" column 
cleaned_bike_df.loc[:, "member_casual"] = cleaned_bike_df["member_casual"].replace({
    "Customer": "casual",
    "Subscriber": "member"
})
```

```python
#use haversine formula which calculates the great circle distance between two points on Earth.
# Function to calculate distance using Haversine formula
def haversine_miles(lat1, lon1, lat2, lon2):
    R = 3958.8  # Earth radius in miles
    
    # Check if any value is missing
    if pd.isnull(lat1) or pd.isnull(lon1) or pd.isnull(lat2) or pd.isnull(lon2):
        return np.nan  # Return NaN if any coordinate is missing

    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    # Haversine formula
    a = np.sin(dlat / 2.0) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0) ** 2
    c = 2 * np.arcsin(np.sqrt(a))

    return R * c  # Distance in miles

cleaned_bike_df = cleaned_bike_df.copy()

# Apply the function to calculate distance for each trip
cleaned_bike_df.loc[:, "distance_traveled_miles"] = cleaned_bike_df.apply(
    lambda row: haversine_miles(row["start_lat"], row["start_lng"], row["end_lat"], row["end_lng"]),
    axis=1
)

cleaned_bike_df
```
